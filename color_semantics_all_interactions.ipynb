```python
%matplotlib inline
from IPython.display import display, display_markdown

import subprocess as sp
import numpy as np
import pandas as pd
import seaborn as sns
import pymc3 as pm
import arviz as az
from bambi import Model

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 8]
plt.rcParams['figure.dpi'] = 300

from subs2vec.utensils import log_timer
from subs2vec.vecs import Vectors
from subs2vec.neighbors import compute_nn

def display_md(md, **kwargs):
    return display_markdown(md, raw=True, **kwargs)

def convert_notebook(title, output='html'):
    convert = sp.run(f'jupyter nbconvert {title}.ipynb --to {output} --output {title}.{output}'.split(' '))
    if convert.returncode == 0:
        display_md(f'Jupyter notebook `{title}` converted successfully.')
    else:
        display_md(f'Error: encountered problem converting Jupyter notebook `{title}`')

def download(fname):
    dl = sp.run(f'wget {fname}'.split(' '))
    if dl.returncode == 0:
        display_md(f'Download of `{fname}` succesful.')
    else:
        display_md(f'Download of `{fname}` failed.')

sns.set(style='whitegrid')
```

# Loading and preparing data

## Loading original participant data (NZ)

```python
df = pd.read_csv('color_semantics_ratings.tsv', sep='\t')
display(df)
```

### Melt dataframe to long format

```python
colors = ['white', 'red', 'orange', 'yellow', 'green', 'blue', 'purple', 'brown', 'black']
df_orig = df.melt(
    id_vars=['group', 'dimension', 'pp_id'],
    value_vars=colors,
    var_name='color',
    value_name='rating',
)
df_orig['experiment'] = 'original'
df_orig['self_vs_other'] = 'self'
display(df_orig)
```

## Loading replication participant data (US)

```python
df_rep = pd.read_csv('replication_subj_pool.csv')
df_rep = df_rep[(df_rep['pp_id'] != 3) & (df_rep['question_type'] != 'catch')]
df_rep = df_rep.drop(columns=['question_type', 'prompt_pre_1'])
df_rep = df_rep.melt(
    id_vars=['dimension', 'color', 'pp_id'],
    value_vars=['value', 'others_choice'],
    var_name='self_vs_other',
    value_name='rating',
)
df_rep['self_vs_other'] = df_rep['self_vs_other'].replace({'value': 'self', 'others_choice': 'other'})
df_rep['group'] = 'sighted'
df_rep['experiment'] = 'replication'

# there is a weird typo in one of the dimensions (?), so let's correct that here:
df_rep['dimension'] = df_rep['dimension'].replace({'like-dis...like': 'like-dislike'})

display(df_rep)
```

## Retrieving vectors for words and dimension word pairs

```python
# as a little bonus, add a 'high-low' dimension
dimension_labels = list(df['dimension'].unique()) + ['high-low']

dimension_pairs = [pair.split('-') for pair in dimension_labels]
from itertools import chain
dimensions = list(chain(*dimension_pairs))
```

```python
# load top 200k vectors (if we load less, we miss some domain vectors)
#vecs = Vectors('../embeddings/subs.en.1e6.vec', n=2e5, d=300, normalize=True)
vecs = Vectors('../embeddings/cc.en.300.vec', n=2e5, d=300, normalize=True)
vecs_dict = vecs.as_dict()

from copy import deepcopy
@log_timer
def filter_vecs(vecs, filter_words):
    filtered_vecs = deepcopy(vecs)
    filtered_vecs.vectors = filtered_vecs.vectors[np.isin(filtered_vecs.words, filter_words)]
    filtered_vecs.words = filtered_vecs.words[np.isin(filtered_vecs.words, filter_words)]
    filtered_vecs.n = len(filtered_vecs.words)
    display_md(f'Filtered {vecs.n} vectors, {filtered_vecs.n} remaining.')
    return filtered_vecs

vecs = filter_vecs(vecs, np.array(colors + dimensions))
color_vecs = filter_vecs(vecs, np.array(colors))
dimension_vecs = filter_vecs(vecs, np.array(dimensions))

dimension_pair_vecs = np.vstack([vecs_dict[pair[0]] - vecs_dict[pair[1]] for pair in dimension_pairs])
```

# Finding color-semantic associations in word embeddings
## Using single dimension words

```python
dimension_neighbors = compute_nn(color_vecs, dimension_vecs.vectors, dimension_vecs.words, num_neighbors=9, whole_matrix=True)
dimension_neighbors = dimension_neighbors.drop(columns=[
    'neighbor -1',
    'neighbor -2',
    'neighbor -3',
    'neighbor -4',
    'neighbor -5',
    'neighbor -6',
    'neighbor -7',
    'neighbor -8',
    'neighbor -9'
]).rename(columns={'target': 'dimension'})
display(dimension_neighbors)
```

## Using dimension axes (word pair contrasts), with nearest neighbor (cosine) method

```python
dimension_neighbors = compute_nn(color_vecs, dimension_pair_vecs, np.array(dimension_labels), num_neighbors=9, whole_matrix=True)
dimension_neighbors = dimension_neighbors.drop(columns=[
    'neighbor -1',
    'neighbor -2',
    'neighbor -3',
    'neighbor -4',
    'neighbor -5',
    'neighbor -6',
    'neighbor -7',
    'neighbor -8',
    'neighbor -9'
]).rename(columns={'target': 'dimension'})
display(dimension_neighbors)
```

It doesn't look like the high-low mapping here corresponds exactly with the one for tone and color in Saysani (2019), but maybe that is to be expected if we do not look specifically for high-_pitch_ and low-_pitch_. (But see section 4!)

# Predicting color semantics association ratings from word embeddings

## Merging data and predictors

### Merge data

```python
df_joint = pd.concat([df_orig, df_rep])
display(df_joint)
```

### Add word frequency (Van Paridon & Thompson, 2020)

```python
freqs = pd.read_csv('../embeddings/dedup.en.words.unigrams.tsv', sep='\t')
freqs['log_freq'] = np.log(freqs['unigram_freq'])
freqs = freqs.drop(columns='unigram_freq')
display(freqs.round(2))
```

```python
df_joint['word1'] = df_joint['dimension'].apply(lambda x: x.split('-')[0])
df_joint['word2'] = df_joint['dimension'].apply(lambda x: x.split('-')[1])
df_joint = df_joint.merge(freqs, left_on='word1', right_on='unigram', how='left')
df_joint = df_joint.merge(freqs, left_on='word2', right_on='unigram', how='left')
df_joint['frequency'] = df_joint['log_freq_x'] - df_joint['log_freq_y']
df_joint = df_joint.drop(columns=[
    'unigram_x',
    'unigram_y',
    'log_freq_x',
    'log_freq_y'
])
display(df_joint.round(2))
```

### Add concreteness (Brysbaert et al., 2014)

```python
concreteness = pd.read_csv('../embeddings/en-brysbaert-2014.tsv', sep='\t')
display(concreteness)
```

```python
df_joint = df_joint.merge(concreteness, left_on='word1', right_on='word', how='left')
df_joint = df_joint.merge(concreteness, left_on='word2', right_on='word', how='left')
df_joint['concreteness'] = df_joint['concreteness_x'] - df_joint['concreteness_y']
df_joint = df_joint.drop(columns=[
    'word_x',
    'word_y',
    'concreteness_x',
    'concreteness_y'
])
display(df_joint.round(2))
```

### Add Small World of Words associations (De Deyne et al., 2018)

```python
swow = pd.read_csv('../embeddings/SWOW-EN.R100.csv')
#swow = swow[(swow['country'] == 'New Zealand') | (swow['country'] == 'Australia')]
#swow = swow[(swow['country'] == 'United States')]
swow = swow[swow['cue'].isin(dimensions)]

# count only R1 (maximal discounting)
swow_R1 = pd.DataFrame(swow.groupby('cue')['R1'].value_counts()).rename(columns={'R1': 'n'})
swow_R1 = swow_R1.reset_index().rename(columns={'R1': 'resp'})
display(swow_R1)

# count R1, R2, and R3 with equal weight (minimal discounting)
swow_all = swow.melt(
    id_vars=['id', 'participantID', 'created_at', 'cue'],
    value_vars=['R1', 'R2', 'R3'],
    value_name='resp',
)
swow_all = pd.DataFrame(swow_all.groupby('cue')['resp'].value_counts()).rename(columns={'resp': 'n'})
swow_all = swow_all.reset_index()
display(swow_all)
```

```python
df_joint = df_joint.merge(swow_all, left_on=['word1', 'color'], right_on=['cue', 'resp'], how='left').fillna(0)
df_joint = df_joint.merge(swow_all, left_on=['word2', 'color'], right_on=['cue', 'resp'], how='left').fillna(0)
df_joint['swow_n'] = df_joint['n_x'] - df_joint['n_y']
df_joint = df_joint.drop(columns=[
    'cue_x',
    'cue_y',
    'resp_x',
    'resp_y',
    'n_x',
    'n_y',
])
display(df_joint)
```

### Add cosine distance (Mikolov et al., 2013)

```python
def norm(x):
    return x / np.linalg.norm(x, 2)
    
df_joint['cosine'] = df_joint.apply(
    lambda x: np.dot(norm(vecs_dict[x['word1']] - vecs_dict[x['word2']]), vecs_dict[x['color']]), axis=1
)
display(df_joint.round(2))
```

### Standardize predictors and plot correlations

```python
def standardize(Series):
    return (Series - Series.mean()) / Series.std()

df_joint['rating_z'] = standardize(df_joint['rating'])
df_joint['frequency_z'] = standardize(df_joint['frequency'])
df_joint['concreteness_z'] = standardize(df_joint['concreteness'])
df_joint['swow_n_z'] = standardize(df_joint['swow_n'])
df_joint['cosine_z'] = standardize(df_joint['cosine'])

df_joint['blind'] = pd.get_dummies(df_joint['group'])['blind']
df_joint['sighted'] = pd.get_dummies(df_joint['group'])['sighted']
df_joint['group_z'] = (df_joint['sighted'] - .5) * 2

df_joint['original'] = pd.get_dummies(df_joint['experiment'])['original']
df_joint['replication'] = pd.get_dummies(df_joint['experiment'])['replication']
df_joint['experiment_z'] = (df_joint['replication'] - .5) * 2

df_joint['other'] = pd.get_dummies(df_joint['self_vs_other'])['other']
df_joint['self_vs_other_z'] = (df_joint['other'] - .5) * 2

display(df_joint)
```

```python
corrs = np.abs(df_joint[[
    'rating',
    'cosine',
    'swow_n',
    'frequency',
    'concreteness',
]].corr()).round(2)
g = sns.heatmap(corrs, vmin=0, vmax=1, annot=True)
```

## Predicting color semantics in original data (blind vs. sighted in NZ)

```python
m_original = Model(df_joint[df_joint['experiment'] == 'original'])
r_original = m_original.fit(
    'rating_z ~ 1 + group_z*frequency_z*concreteness_z*cosine_z*swow_n_z',
    random=[
        '1 + frequency_z*concreteness_z*cosine_z*swow_n_z + color + dimension|pp_id',
        '1 + group_z|dimension',
        '1 + group_z|color',
    ],
    init='advi+adapt_diag',
    chains=3,
    samples=2000,
    tune=2000,
    target_accept=.9,
)
```

```python
varnames = [varname for varname in m_original.backend.trace.varnames if (('|' not in varname)
                                                                         and ('_sd' not in varname))]
g = az.plot_forest(r_original, combined=True, hdi_prob=.95,
                   figsize=[9, .6 + len(varnames) * .3],
                   var_names=varnames)
g[0].axvline(0, color='.8', linewidth=2);
```

In the original experiment, the main effects of cosine and SWOW look pretty robust.

There is no difference in the effect of SWOW between the sighted and blind participants, but there is a non-zero group:cosine interaction, meaning that word embeddings are more predictive of sighted participants' ratings than of blind participants' ratings. In fact, if we compute the conditional effect of cosine in each of the groups (which we can do post-hoc, see figure below) it seems that in the blind group the effect of cosine might well be zero.

```python
# compute and plot conditional effect of cosine in sighted and blind groups
from copy import deepcopy
trace_orig = deepcopy(m_original.backend.trace)
trace_orig.add_values({
    'blind:cosine_z': trace_orig['cosine_z'] - trace_orig['group_z:cosine_z'],
    'sighted:cosine_z': trace_orig['cosine_z'] + trace_orig['group_z:cosine_z']
})

varnames = ['blind:cosine_z', 'sighted:cosine_z']
g = az.plot_forest(trace_orig, combined=True, hdi_prob=.95,
                   figsize=[9, .6 + len(varnames) * .3],
                   var_names=varnames)
g[0].axvline(0, color='.8', linewidth=2);
```

## Predicting color semantics in replication data (sighted in US)

```python
m_replication = Model(df_joint[(df_joint['experiment'] == 'replication') & (df_joint['self_vs_other'] == 'self')])
r_replication = m_replication.fit(
    'rating_z ~ 1 + frequency_z*concreteness_z*cosine_z*swow_n_z',
    random=[
        '1 + frequency_z*concreteness_z*cosine_z*swow_n_z + color + dimension|pp_id',
        '1|dimension',
        '1|color',
    ],
    init='advi+adapt_diag',
    chains=3,
    samples=2000,
    tune=2000,
    target_accept=.9,
)
```

```python
varnames = [varname for varname in m_replication.backend.trace.varnames if (('|' not in varname)
                                                                         and ('_sd' not in varname))]
g = az.plot_forest(r_replication, combined=True, hdi_prob=.95,
                   figsize=[9, .6 + len(varnames) * .3],
                   var_names=varnames)
g[0].axvline(0, color='.8', linewidth=2);
```

In the replication model we see a fairly robust effect of cosine and a slightly smaller effect of SWOW. These observations findings are broadly consistent with what we see in the sighted group in the original experiment.

## Differences in color semantics between original and replication data

```python
m_orig_vs_rep = Model(df_joint[df_joint['self_vs_other'] == 'self'])
r_orig_vs_rep = m_orig_vs_rep.fit(
    'rating_z ~ 1'
    + ' + group_z*frequency_z*concreteness_z*cosine_z*swow_n_z'
    + ' + experiment_z*frequency_z*concreteness_z*cosine_z*swow_n_z',
    random=[
        '1 + frequency_z*concreteness_z*cosine_z*swow_n_z + color + dimension|pp_id',
        '1 + experiment_z + group_z|dimension',
        '1 + experiment_z + group_z|color',
    ],
    init='advi+adapt_diag',
    chains=3,
    samples=2000,
    tune=2000,
    target_accept=.9,
)
```

```python
varnames = [varname for varname in m_orig_vs_rep.backend.trace.varnames if (('|' not in varname)
                                                                         and ('_sd' not in varname))]
g = az.plot_forest(r_orig_vs_rep, combined=True, hdi_prob=.95,
                   figsize=[9, .6 + len(varnames) * .3],
                   var_names=varnames)
g[0].axvline(0, color='.8', linewidth=2);
```

When pooling the original and replication experiments into a single model, the main effects of cosine and SWOW look fairly robust.

This pooled model also shows the group:cosine interaction, meaning that word embeddings are _more_ predictive of sighted participants' ratings than of blind participants' ratings. If we compute the conditional effect of cosine in each of the groups (which we can do post-hoc, see figure below) like we did when modelling the original experiment, it seems that in the blind group the effect of cosine is highly uncertain and quite possibly zero.

The effect of SWOW is similar in direction to what we see when we model the original and replication experiments separately, but there is a small experiment:swow interaction effect, meaning SWOW is more predictive of ratings in sighted participants in the NZ sample than in the US sample. It's hard to say what the origin of this effect would be, because the NZ participants' linguistic associations stem from a smaller and more isolated language community, which one might expect to cause them to be _more_ different from the SWOW associations rather than _less_ different. (This is pretty speculative, though.)

```python
# compute and plot conditional effect of cosine in sighted and blind groups
from copy import deepcopy
trace_vs = deepcopy(m_orig_vs_rep.backend.trace)
trace_vs.add_values({
    # effect of cosine in blind group in orginal experiment
    'blind:cosine_z': trace_vs['cosine_z'] - trace_vs['group_z:cosine_z'] - trace_vs['experiment_z:cosine_z'],
    # effect of cosine in sighted group across both experiments
    'sighted:cosine_z': trace_vs['cosine_z'] + trace_vs['group_z:cosine_z'],
})

varnames = ['blind:cosine_z', 'sighted:cosine_z']
g = az.plot_forest(trace_vs, combined=True, hdi_prob=.95,
                   figsize=[9, .6 + len(varnames) * .3],
                   var_names=varnames)
g[0].axvline(0, color='.8', linewidth=2);
```

## Differences in color semantics between self-ratings and other-ratings

```python
m_self_vs_other = Model(df_joint[df_joint['experiment'] == 'replication'])
r_self_vs_other = m_self_vs_other.fit(
    'rating_z ~ 1'
    + ' + self_vs_other_z*frequency_z*concreteness_z*cosine_z*swow_n_z',
    random=[
        '1 + frequency_z*concreteness_z*cosine_z*swow_n_z + color + dimension|pp_id',
        '1 + self_vs_other_z|dimension',
        '1 + self_vs_other_z|color',
    ],
    init='advi+adapt_diag',
    chains=3,
    samples=2000,
    tune=2000,
    target_accept=.9,
)
```

```python
varnames = [varname for varname in m_self_vs_other.backend.trace.varnames if (('|' not in varname)
                                                                         and ('_sd' not in varname))]
g = az.plot_forest(r_self_vs_other, combined=True, hdi_prob=.95,
                   figsize=[9, .6 + len(varnames) * .3],
                   var_names=varnames)
g[0].axvline(0, color='.8', linewidth=2);
```

There appears to be no discernable difference between self- and other-ratings in the replication experiment.

### Self-other differences

```python
df_self = df_joint[(df_joint['experiment'] == 'replication') & (df_joint['self_vs_other'] == 'self')].sort_values(['pp_id', 'dimension', 'color'])
df_other = df_joint[df_joint['self_vs_other'] == 'other'].sort_values(['pp_id', 'dimension', 'color'])

df_self['rating_diff'] = np.abs(df_self['rating'].values - df_other['rating'].values)
df_self['rating_diff_z'] = standardize(df_self['rating_diff'])
```

```python
m_diff = Model(df_self)
r_diff = m_diff.fit(
    'rating_diff_z ~ 1 + frequency_z*concreteness_z*cosine_z*swow_n_z',
    random=[
        '1 + frequency_z*concreteness_z*cosine_z*swow_n_z + color + dimension|pp_id',
        '1|dimension',
        '1|color',
    ],
    init='advi+adapt_diag',
    chains=3,
    samples=2000,
    tune=2000,
    target_accept=.9,
)
```

```python
varnames = [varname for varname in m_diff.backend.trace.varnames if (('|' not in varname)
                                                                         and ('_sd' not in varname))]
g = az.plot_forest(r_diff, combined=True, hdi_prob=.95,
                   figsize=[9, .6 + len(varnames) * .3],
                   var_names=varnames)
g[0].axvline(0, color='.8', linewidth=2);
```
